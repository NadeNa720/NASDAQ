```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tickers = ["AAPL", "MSFT", "GOOGL", "AMZN", "TSLA"]

periods = [
    ("2024-01-01", "2024-11-01"),
    ("2023-01-01", "2024-01-01"),
    ("2022-01-01", "2023-01-01"),
    ("2021-01-01", "2022-01-01"),
    ("2020-01-01", "2021-01-01")
]
```

```python
dataframes = {}

for ticker, (start, end) in zip(tickers, periods):
    data = yf.Ticker(ticker).history(start=start, end=end, interval="1d")
    
    df = pd.DataFrame({
        "Date": data.index,
        "Close": data["Close"].values
    })
    
    df.reset_index(drop=True, inplace=True)
    
    dataframes[ticker] = df
```

```python
for ticker, df in dataframes.items():
    print(f"\n--- {ticker} ---")
    
    print("Descriptive statistics:")
    print(df.describe())
    
    print("\nKPI:")
    print("Completeness:\n", df.isnull().sum())
  
```

```python
for ticker, df in dataframes.items():
    df.plot(x="Date", y="Close", title=f"{ticker} Closing Prices")
    plt.show()
```
